```python
import pandas as pd
import numpy as np
from tkinter import *
from tkinter.ttk import *
from pandastable import Table, TableModel
```

### ***Leitura e Tratamento do Dataset de Covid para Aplicação das Técnicas de Anonimização.***

```python
df_covid = pd.read_csv('covid.csv', sep=';')
df_covid.drop(columns='Unnamed: 0', inplace=True)
#-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_
df_covid['CodigoMunicipio'] = df_covid['CodigoMunicipio'].astype(int)
df_covid.head()
```

### ***Leitura e Tratamento do Dataset de Regiões para Mesclagem e Aplicação das Técnicas de Anonimização.***

```python
df_regioes = pd.read_csv('regioes.csv')
df_regioes.drop(columns='Unnamed: 0', inplace=True)
df_regioes.rename(columns={'Código IBGE do Município': 'CodigoMunicipio'}, inplace=True)
df_regioes.head()
```

### ***Junção dos 2 Datasets para realizar as Generalizações.***

```python
df_covid_and_regioes = pd.merge(df_covid, df_regioes, on="CodigoMunicipio", how="outer")
df_covid_and_regioes.head()
```

### ***Função de Anonimização da Coluna 'Municipio'.***

```python
for cod_muni in df_covid_and_regioes.CodigoMunicipio.unique():
    df_covid_and_regioes['Municipio'] = df_covid_and_regioes['Nome da Mesoregião']
```

```python
df_covid_and_regioes_new = df_covid_and_regioes.drop(columns=['Nome da Mesoregião','Sigla da Unidade da Federação','Nome das Grandes Regiões'])
df_covid_and_regioes_new
```

### ***Exportando o Dataset Anonimizado 01 Conforme Solicitado no Formato CSV.***

```python
df_covid_and_regioes_new.to_csv('covid_anon01.csv', index=None)
```

# Programa de Visualização dos Dados

```python
class NewWindow(Toplevel):     
    def __init__(self, master = None, df=''):
         
        super().__init__(master = master)
        my_df = pd.read_csv(df)
        self.main = self.master
        self.main.geometry('600x400+200+100')
        self.main.title('Exibir datasets anonimalizados')
        f = Label(self, text ="This is a new Window")
        f.pack(fill=BOTH,expand=1)
        self.table = pt = Table(f, dataframe=my_df,
                                showtoolbar=True, showstatusbar=True)
        pt.show()
        return 
    
master = Tk()
master.title('Datasets Anonimalizados')
master.geometry('600x400+200+100')
 
label = Label(master, text ="Pressione um botão para exibir um Dataset")
label.pack(side = TOP, pady = 10)

btn1 = Button(master, text ="covid_anon01.csv")
btn1.bind("<Button>", lambda e: NewWindow(master, df='covid_anon01.csv'))
btn1.pack(side = TOP, pady = 10)


btn2 = Button(master, text ="covid_anon02.csv")
btn2.bind("<Button>", lambda e: NewWindow(master, df='covid_anon02.csv'))
btn2.pack(side = TOP, pady = 10)


btn3 = Button(master, text ="covid_anon03.csv")
btn3.bind("<Button>", lambda e: NewWindow(master, df='covid_anon03.csv'))
btn3.pack(side = TOP, pady = 10)


btn4 = Button(master, text ="covid_anon04.csv")
btn4.bind("<Button>", lambda e: NewWindow(master, df='covid_anon04.csv'))
btn4.pack(side = TOP, pady = 10)

mainloop()
```
